```python
# Import libraries
import random
import math
# Tanh activation function
def tanh(x):
    return math.tanh(x)
# Forward propagation function
def forward_pass(i1, i2, b1, b2):
 # Initialize random weights in [-0.5, 0.5]
    w = [random.uniform(-0.5, 0.5) for _ in range(8)]
    w1, w2, w3, w4, w5, w6, w7, w8 = w
 # Hidden layer
    net_h1 = i1*w1 + i2*w2 + b1
    net_h2 = i1*w3 + i2*w4 + b1
    h1 = tanh(net_h1)
    h2 = tanh(net_h2)
    # Output layer
    net_o1 = h1*w5 + h2*w6 + b2
    net_o2 = h1*w7 + h2*w8 + b2
    o1 = tanh(net_o1)
    o2 = tanh(net_o2)
    return {
        "weights": w,
        "net_h1": net_h1,
        "net_h2": net_h2,
        "net_o1": net_o1,
        "net_o2": net_o2,
        "h1": h1,
        "h2": h2,
        "o1": o1,
        "o2": o2
    }
# Main execution
if __name__ == "__main__":
# Inputs and biases
    i1, i2 = 0.05, 0.10
    b1 = 0.5
    b2 = 0.7
    result = forward_pass(i1, i2, b1, b2)
    print("Final Outputs:")
    print("o1 =", result["o1"])
    print("o2 =", result["o2"])
```
